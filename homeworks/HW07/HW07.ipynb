```python
# Домашняя работа 7: Кластеризация

# 2.3.1. Загрузка данных и первичный анализ

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import json
import warnings
warnings.filterwarnings('ignore')

# Устанавливаем стиль графиков
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# Создаем необходимые папки
artifacts_dir = Path("artifacts")
artifacts_dir.mkdir(exist_ok=True)
(artifacts_dir / "labels").mkdir(exist_ok=True)
(artifacts_dir / "figures").mkdir(exist_ok=True)

# Загружаем данные
data_dir = Path("data")
datasets = {
    "ds01": "S07-hw-dataset-01.csv",
    "ds02": "S07-hw-dataset-02.csv", 
    "ds03": "S07-hw-dataset-03.csv"
}

# Словарь для хранения данных
dfs = {}
sample_ids = {}
X_raw = {}

for name, filename in datasets.items():
    filepath = data_dir / filename
    df = pd.read_csv(filepath)
    dfs[name] = df
    sample_ids[name] = df['sample_id']
    X_raw[name] = df.drop(columns=['sample_id'])
    
    print(f"\n{'='*60}")
    print(f"Датасет: {filename}")
    print(f"{'='*60}")
    
    print(f"\nПервые 5 строк:")
    print(df.head())
    
    print(f"\nИнформация о датасете:")
    print(df.info())
    
    print(f"\nБазовые статистики:")
    print(df.describe())
    
    print(f"\nПроверка пропусков:")
    missing = df.isnull().sum()
    missing_pct = (missing / len(df)) * 100
    missing_df = pd.DataFrame({
        'Количество пропусков': missing,
        'Доля пропусков (%)': missing_pct
    })
    print(missing_df[missing_df['Количество пропусков'] > 0])
    
    print(f"\nТипы признаков:")
    for col in df.columns:
        if col != 'sample_id':
            print(f"  {col}: {df[col].dtype} ({'числовой' if np.issubdtype(df[col].dtype, np.number) else 'категориальный'})")

# 2.3.2. Препроцессинг

from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
import warnings
warnings.filterwarnings('ignore')

# Создаем пайплайны препроцессинга для каждого датасета
preprocessors = {}

# Для ds01: только числовые признаки, масштабирование
preprocessors['ds01'] = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),  # На всякий случай
    ('scaler', StandardScaler())
])

# Для ds02: только числовые признаки, масштабирование
preprocessors['ds02'] = Pipeline([
    ('scaler', StandardScaler())
])

# Для ds03: только числовые признаки, масштабирование  
preprocessors['ds03'] = Pipeline([
    ('scaler', StandardScaler())
])

# Применяем препроцессинг
X_processed = {}
for name in datasets.keys():
    X_processed[name] = preprocessors[name].fit_transform(X_raw[name])
    print(f"\nДатасет {name}:")
    print(f"  Исходная форма: {X_raw[name].shape}")
    print(f"  После препроцессинга: {X_processed[name].shape}")

# 2.3.3. Модели кластеризации

from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from scipy.cluster.hierarchy import dendrogram, linkage

# Словари для хранения результатов
results = {}
best_configs = {}

# Датасет 01: Числовые признаки в разных шкалах + шум

# Эксперимент с KMeans
print("Датасет 01: Подбор k для KMeans")
print("-" * 50)

silhouette_scores = []
db_scores = []
ch_scores = []
k_range = range(2, 21)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_processed['ds01'])
    
    # Вычисляем метрики (исключаем случаи, когда все точки в одном кластере)
    if len(np.unique(labels)) > 1:
        silhouette = silhouette_score(X_processed['ds01'], labels)
        db = davies_bouldin_score(X_processed['ds01'], labels)
        ch = calinski_harabasz_score(X_processed['ds01'], labels)
    else:
        silhouette = -1
        db = np.inf
        ch = 0
    
    silhouette_scores.append(silhouette)
    db_scores.append(db)
    ch_scores.append(ch)
    
    print(f"k={k:2d}: Silhouette={silhouette:.4f}, DB={db:.4f}, CH={ch:.2f}")

# Визуализация
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Silhouette
axes[0].plot(k_range, silhouette_scores, marker='o', linewidth=2)
axes[0].set_xlabel('Количество кластеров (k)')
axes[0].set_ylabel('Silhouette Score')
axes[0].set_title('Silhouette Score vs k')
axes[0].grid(True, alpha=0.3)

# Davies-Bouldin
axes[1].plot(k_range, db_scores, marker='o', linewidth=2, color='orange')
axes[1].set_xlabel('Количество кластеров (k)')
axes[1].set_ylabel('Davies-Bouldin Index')
axes[1].set_title('Davies-Bouldin Index vs k')
axes[1].grid(True, alpha=0.3)

# Calinski-Harabasz
axes[2].plot(k_range, ch_scores, marker='o', linewidth=2, color='green')
axes[2].set_xlabel('Количество кластеров (k)')
axes[2].set_ylabel('Calinski-Harabasz Score')
axes[2].set_title('Calinski-Harabasz Score vs k')
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('artifacts/figures/ds01_kmeans_metrics.png', dpi=150, bbox_inches='tight')
plt.show()

# Выбираем оптимальное k по silhouette
best_k_idx = np.argmax(silhouette_scores)
best_k = k_range[best_k_idx]
print(f"\nОптимальное k по silhouette: {best_k}")

# Обучаем KMeans с оптимальным k
kmeans_best = KMeans(n_clusters=best_k, random_state=42, n_init=10)
kmeans_labels = kmeans_best.fit_predict(X_processed['ds01'])

# Сохраняем результаты
results['ds01'] = {
    'KMeans': {
        'labels': kmeans_labels,
        'k': int(best_k),  # Преобразуем в int для JSON
        'silhouette': float(silhouette_scores[best_k_idx]),
        'davies_bouldin': float(db_scores[best_k_idx]),
        'calinski_harabasz': float(ch_scores[best_k_idx]),
        'params': {'n_clusters': int(best_k), 'random_state': 42}
    }
}

# Для датасета 01 пробуем DBSCAN
print("\nДатасет 01: Подбор параметров для DBSCAN")
print("-" * 50)

# Исследуем различные значения eps и min_samples
eps_values = np.linspace(0.1, 2.0, 20)
min_samples_values = [2, 3, 5, 10]

dbscan_results = []
for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(X_processed['ds01'])
        
        n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
        noise_ratio = np.sum(labels == -1) / len(labels) if len(labels) > 0 else 0
        
        if n_clusters > 1:
            # Вычисляем метрики только на non-noise точках
            non_noise_mask = labels != -1
            if np.sum(non_noise_mask) > 1 and len(np.unique(labels[non_noise_mask])) > 1:
                silhouette = silhouette_score(X_processed['ds01'][non_noise_mask], 
                                            labels[non_noise_mask])
                db = davies_bouldin_score(X_processed['ds01'][non_noise_mask], 
                                        labels[non_noise_mask])
                ch = calinski_harabasz_score(X_processed['ds01'][non_noise_mask], 
                                           labels[non_noise_mask])
            else:
                silhouette = -1
                db = np.inf
                ch = 0
        else:
            silhouette = -1
            db = np.inf
            ch = 0
        
        dbscan_results.append({
            'eps': float(eps),
            'min_samples': int(min_samples),
            'n_clusters': int(n_clusters),
            'noise_ratio': float(noise_ratio),
            'silhouette': float(silhouette),
            'davies_bouldin': float(db),
            'calinski_harabasz': float(ch)
        })

# Преобразуем в DataFrame для анализа
dbscan_df = pd.DataFrame(dbscan_results)

# Фильтруем разумные результаты (не слишком много шума, не слишком мало кластеров)
reasonable_results = dbscan_df[
    (dbscan_df['noise_ratio'] < 0.3) & 
    (dbscan_df['n_clusters'] > 1) & 
    (dbscan_df['n_clusters'] < 20)
]

if len(reasonable_results) > 0:
    # Лучший результат по silhouette
    best_idx = reasonable_results['silhouette'].idxmax()
    best_result = reasonable_results.loc[best_idx]
    
    print(f"Лучшие параметры DBSCAN:")
    print(f"  eps: {best_result['eps']:.3f}")
    print(f"  min_samples: {best_result['min_samples']}")
    print(f"  Количество кластеров: {best_result['n_clusters']}")
    print(f"  Доля шума: {best_result['noise_ratio']:.3f}")
    print(f"  Silhouette: {best_result['silhouette']:.4f}")
    
    # Обучаем DBSCAN с лучшими параметрами
    dbscan_best = DBSCAN(eps=best_result['eps'], 
                        min_samples=int(best_result['min_samples']))
    dbscan_labels = dbscan_best.fit_predict(X_processed['ds01'])
    
    # Вычисляем метрики для non-noise точек
    non_noise_mask = dbscan_labels != -1
    if np.sum(non_noise_mask) > 1 and len(np.unique(dbscan_labels[non_noise_mask])) > 1:
        silhouette = silhouette_score(X_processed['ds01'][non_noise_mask], 
                                    dbscan_labels[non_noise_mask])
        db = davies_bouldin_score(X_processed['ds01'][non_noise_mask], 
                                dbscan_labels[non_noise_mask])
        ch = calinski_harabasz_score(X_processed['ds01'][non_noise_mask], 
                                   dbscan_labels[non_noise_mask])
    else:
        silhouette = -1
        db = np.inf
        ch = 0
    
    results['ds01']['DBSCAN'] = {
        'labels': dbscan_labels,
        'eps': float(best_result['eps']),
        'min_samples': int(best_result['min_samples']),
        'n_clusters': int(best_result['n_clusters']),
        'noise_ratio': float(best_result['noise_ratio']),
        'silhouette': float(silhouette),
        'davies_bouldin': float(db),
        'calinski_harabasz': float(ch),
        'params': {'eps': float(best_result['eps']), 
                  'min_samples': int(best_result['min_samples'])}
    }
else:
    print("Не удалось найти разумные параметры для DBSCAN")

# Визуализация silhouette в зависимости от eps для фиксированного min_samples
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

for idx, min_samples in enumerate(min_samples_values[:4]):
    subset = dbscan_df[dbscan_df['min_samples'] == min_samples]
    ax = axes[idx]
    
    for eps_val in eps_values:
        eps_data = subset[subset['eps'] == float(eps_val)]
        if len(eps_data) > 0:
            ax.scatter(eps_val, eps_data['silhouette'].values[0], 
                      c='blue', alpha=0.6, s=50)
    
    ax.set_xlabel('eps')
    ax.set_ylabel('Silhouette Score')
    ax.set_title(f'min_samples = {min_samples}')
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('artifacts/figures/ds01_dbscan_silhouette.png', dpi=150, bbox_inches='tight')
plt.show()

# Датасет 02: Нелинейная структура + выбросы

print("\n" + "="*60)
print("Датасет 02: Анализ")
print("="*60)

# KMeans для датасета 02
print("\nДатасет 02: Подбор k для KMeans")
print("-" * 50)

silhouette_scores_ds02 = []
db_scores_ds02 = []
ch_scores_ds02 = []
k_range_ds02 = range(2, 11)

for k in k_range_ds02:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_processed['ds02'])
    
    if len(np.unique(labels)) > 1:
        silhouette = silhouette_score(X_processed['ds02'], labels)
        db = davies_bouldin_score(X_processed['ds02'], labels)
        ch = calinski_harabasz_score(X_processed['ds02'], labels)
    else:
        silhouette = -1
        db = np.inf
        ch = 0
    
    silhouette_scores_ds02.append(silhouette)
    db_scores_ds02.append(db)
    ch_scores_ds02.append(ch)
    
    print(f"k={k:2d}: Silhouette={silhouette:.4f}, DB={db:.4f}, CH={ch:.2f}")

# Визуализация
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(k_range_ds02, silhouette_scores_ds02, marker='o', linewidth=2)
ax.set_xlabel('Количество кластеров (k)')
ax.set_ylabel('Silhouette Score')
ax.set_title('Датасет 02: Silhouette Score vs k (KMeans)')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('artifacts/figures/ds02_kmeans_silhouette.png', dpi=150, bbox_inches='tight')
plt.show()

# Выбираем оптимальное k
best_k_idx_ds02 = np.argmax(silhouette_scores_ds02)
best_k_ds02 = k_range_ds02[best_k_idx_ds02]
print(f"\nОптимальное k по silhouette: {best_k_ds02}")

# Обучаем KMeans
kmeans_best_ds02 = KMeans(n_clusters=best_k_ds02, random_state=42, n_init=10)
kmeans_labels_ds02 = kmeans_best_ds02.fit_predict(X_processed['ds02'])

# Сохраняем результаты
results['ds02'] = {
    'KMeans': {
        'labels': kmeans_labels_ds02,
        'k': int(best_k_ds02),
        'silhouette': float(silhouette_scores_ds02[best_k_idx_ds02]),
        'davies_bouldin': float(db_scores_ds02[best_k_idx_ds02]),
        'calinski_harabasz': float(ch_scores_ds02[best_k_idx_ds02]),
        'params': {'n_clusters': int(best_k_ds02), 'random_state': 42}
    }
}

# Для датасета 02 пробуем Agglomerative Clustering

print("\nДатасет 02: Agglomerative Clustering с разными linkage")
print("-" * 50)

linkage_methods = ['ward', 'complete', 'average', 'single']
best_silhouette = -1
best_linkage = None
best_labels = None
best_k = None

for linkage_method in linkage_methods:
    print(f"\nLinkage: {linkage_method}")
    print("-" * 30)
    
    silhouette_scores_linkage = []
    
    for k in range(2, 11):
        agg = AgglomerativeClustering(n_clusters=k, linkage=linkage_method)
        labels = agg.fit_predict(X_processed['ds02'])
        
        if len(np.unique(labels)) > 1:
            silhouette = silhouette_score(X_processed['ds02'], labels)
        else:
            silhouette = -1
        
        silhouette_scores_linkage.append(silhouette)
        print(f"  k={k}: Silhouette={silhouette:.4f}")
    
    # Находим лучший k для этого linkage
    best_k_idx = np.argmax(silhouette_scores_linkage)
    current_best_silhouette = silhouette_scores_linkage[best_k_idx]
    
    if current_best_silhouette > best_silhouette:
        best_silhouette = current_best_silhouette
        best_linkage = linkage_method
        best_k = best_k_idx + 2  # +2 потому что range начинается с 2
        
        # Обучаем с лучшими параметрами
        agg_best = AgglomerativeClustering(n_clusters=best_k, linkage=best_linkage)
        best_labels = agg_best.fit_predict(X_processed['ds02'])

print(f"\nЛучшие параметры:")
print(f"  Linkage: {best_linkage}")
print(f"  k: {best_k}")
print(f"  Silhouette: {best_silhouette:.4f}")

# Вычисляем полный набор метрик
if best_labels is not None and len(np.unique(best_labels)) > 1:
    silhouette = silhouette_score(X_processed['ds02'], best_labels)
    db = davies_bouldin_score(X_processed['ds02'], best_labels)
    ch = calinski_harabasz_score(X_processed['ds02'], best_labels)
    
    results['ds02']['Agglomerative'] = {
        'labels': best_labels,
        'k': int(best_k),
        'linkage': best_linkage,
        'silhouette': float(silhouette),
        'davies_bouldin': float(db),
        'calinski_harabasz': float(ch),
        'params': {'n_clusters': int(best_k), 'linkage': best_linkage}
    }

# Визуализация сравнения linkage
fig, ax = plt.subplots(figsize=(10, 6))
for linkage_method in linkage_methods:
    silhouette_scores = []
    for k in range(2, 11):
        agg = AgglomerativeClustering(n_clusters=k, linkage=linkage_method)
        labels = agg.fit_predict(X_processed['ds02'])
        if len(np.unique(labels)) > 1:
            silhouette = silhouette_score(X_processed['ds02'], labels)
        else:
            silhouette = -1
        silhouette_scores.append(silhouette)
    
    ax.plot(range(2, 11), silhouette_scores, marker='o', linewidth=2, 
            label=f'{linkage_method}', alpha=0.7)

ax.set_xlabel('Количество кластеров (k)')
ax.set_ylabel('Silhouette Score')
ax.set_title('Датасет 02: Сравнение linkage методов (Agglomerative)')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('artifacts/figures/ds02_agglomerative_linkage.png', dpi=150, bbox_inches='tight')
plt.show()

# Датасет 03: Кластеры разной плотности + шум

print("\n" + "="*60)
print("Датасет 03: Анализ")
print("="*60)

# KMeans для датасета 03
print("\nДатасет 03: Подбор k для KMeans")
print("-" * 50)

silhouette_scores_ds03 = []
db_scores_ds03 = []
ch_scores_ds03 = []
k_range_ds03 = range(2, 15)

for k in k_range_ds03:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_processed['ds03'])
    
    if len(np.unique(labels)) > 1:
        silhouette = silhouette_score(X_processed['ds03'], labels)
        db = davies_bouldin_score(X_processed['ds03'], labels)
        ch = calinski_harabasz_score(X_processed['ds03'], labels)
    else:
        silhouette = -1
        db = np.inf
        ch = 0
    
    silhouette_scores_ds03.append(silhouette)
    db_scores_ds03.append(db)
    ch_scores_ds03.append(ch)
    
    print(f"k={k:2d}: Silhouette={silhouette:.4f}, DB={db:.4f}, CH={ch:.2f}")

# Визуализация
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(k_range_ds03, silhouette_scores_ds03, marker='o', linewidth=2)
ax.set_xlabel('Количество кластеров (k)')
ax.set_ylabel('Silhouette Score')
ax.set_title('Датасет 03: Silhouette Score vs k (KMeans)')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('artifacts/figures/ds03_kmeans_silhouette.png', dpi=150, bbox_inches='tight')
plt.show()

# Выбираем оптимальное k
best_k_idx_ds03 = np.argmax(silhouette_scores_ds03)
best_k_ds03 = k_range_ds03[best_k_idx_ds03]
print(f"\nОптимальное k по silhouette: {best_k_ds03}")

# Обучаем KMeans
kmeans_best_ds03 = KMeans(n_clusters=best_k_ds03, random_state=42, n_init=10)
kmeans_labels_ds03 = kmeans_best_ds03.fit_predict(X_processed['ds03'])

# Сохраняем результаты
results['ds03'] = {
    'KMeans': {
        'labels': kmeans_labels_ds03,
        'k': int(best_k_ds03),
        'silhouette': float(silhouette_scores_ds03[best_k_idx_ds03]),
        'davies_bouldin': float(db_scores_ds03[best_k_idx_ds03]),
        'calinski_harabasz': float(ch_scores_ds03[best_k_idx_ds03]),
        'params': {'n_clusters': int(best_k_ds03), 'random_state': 42}
    }
}

# Для датасета 03 пробуем DBSCAN (идеально для разной плотности)

print("\nДатасет 03: Подбор параметров для DBSCAN")
print("-" * 50)

# Для датасета с разной плотностью пробуем разные параметры
eps_values_ds03 = np.linspace(0.1, 1.0, 20)
min_samples_values_ds03 = [2, 3, 5, 7, 10]

dbscan_results_ds03 = []
for eps in eps_values_ds03:
    for min_samples in min_samples_values_ds03:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(X_processed['ds03'])
        
        n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
        noise_ratio = np.sum(labels == -1) / len(labels) if len(labels) > 0 else 0
        
        if n_clusters > 1:
            non_noise_mask = labels != -1
            if np.sum(non_noise_mask) > 1 and len(np.unique(labels[non_noise_mask])) > 1:
                silhouette = silhouette_score(X_processed['ds03'][non_noise_mask], 
                                            labels[non_noise_mask])
                db = davies_bouldin_score(X_processed['ds03'][non_noise_mask], 
                                        labels[non_noise_mask])
                ch = calinski_harabasz_score(X_processed['ds03'][non_noise_mask], 
                                           labels[non_noise_mask])
            else:
                silhouette = -1
                db = np.inf
                ch = 0
        else:
            silhouette = -1
            db = np.inf
            ch = 0
        
        dbscan_results_ds03.append({
            'eps': float(eps),
            'min_samples': int(min_samples),
            'n_clusters': int(n_clusters),
            'noise_ratio': float(noise_ratio),
            'silhouette': float(silhouette),
            'davies_bouldin': float(db),
            'calinski_harabasz': float(ch)
        })

dbscan_df_ds03 = pd.DataFrame(dbscan_results_ds03)

# Ищем лучшие параметры
reasonable_results_ds03 = dbscan_df_ds03[
    (dbscan_df_ds03['noise_ratio'] < 0.4) &  # Допускаем больше шума для такого датасета
    (dbscan_df_ds03['n_clusters'] > 1) & 
    (dbscan_df_ds03['n_clusters'] < 10)
]

if len(reasonable_results_ds03) > 0:
    # Сначала сортируем по silhouette, затем по меньшему количеству шума
    sorted_results = reasonable_results_ds03.sort_values(
        ['silhouette', 'noise_ratio'], 
        ascending=[False, True]
    )
    
    best_result_ds03 = sorted_results.iloc[0]
    
    print(f"Лучшие параметры DBSCAN:")
    print(f"  eps: {best_result_ds03['eps']:.3f}")
    print(f"  min_samples: {best_result_ds03['min_samples']}")
    print(f"  Количество кластеров: {best_result_ds03['n_clusters']}")
    print(f"  Доля шума: {best_result_ds03['noise_ratio']:.3f}")
    print(f"  Silhouette: {best_result_ds03['silhouette']:.4f}")
    
    # Обучаем DBSCAN
    dbscan_best_ds03 = DBSCAN(eps=best_result_ds03['eps'], 
                             min_samples=int(best_result_ds03['min_samples']))
    dbscan_labels_ds03 = dbscan_best_ds03.fit_predict(X_processed['ds03'])
    
    # Вычисляем метрики
    non_noise_mask = dbscan_labels_ds03 != -1
    if np.sum(non_noise_mask) > 1 and len(np.unique(dbscan_labels_ds03[non_noise_mask])) > 1:
        silhouette = silhouette_score(X_processed['ds03'][non_noise_mask], 
                                    dbscan_labels_ds03[non_noise_mask])
        db = davies_bouldin_score(X_processed['ds03'][non_noise_mask], 
                                dbscan_labels_ds03[non_noise_mask])
        ch = calinski_harabasz_score(X_processed['ds03'][non_noise_mask], 
                                   dbscan_labels_ds03[non_noise_mask])
    else:
        silhouette = -1
        db = np.inf
        ch = 0
    
    results['ds03']['DBSCAN'] = {
        'labels': dbscan_labels_ds03,
        'eps': float(best_result_ds03['eps']),
        'min_samples': int(best_result_ds03['min_samples']),
        'n_clusters': int(best_result_ds03['n_clusters']),
        'noise_ratio': float(best_result_ds03['noise_ratio']),
        'silhouette': float(silhouette),
        'davies_bouldin': float(db),
        'calinski_harabasz': float(ch),
        'params': {'eps': float(best_result_ds03['eps']), 
                  'min_samples': int(best_result_ds03['min_samples'])}
    }

# Визуализация silhouette vs eps для датасета 03
fig, ax = plt.subplots(figsize=(10, 6))

for min_samples in [2, 5, 10]:
    subset = dbscan_df_ds03[dbscan_df_ds03['min_samples'] == min_samples]
    if len(subset) > 0:
        ax.plot(subset['eps'], subset['silhouette'], 
                marker='o', linewidth=2, label=f'min_samples={min_samples}', alpha=0.7)

ax.set_xlabel('eps')
ax.set_ylabel('Silhouette Score')
ax.set_title('Датасет 03: Silhouette Score vs eps (DBSCAN)')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('artifacts/figures/ds03_dbscan_silhouette.png', dpi=150, bbox_inches='tight')
plt.show()

# 2.3.4. Сводка метрик

print("\n" + "="*60)
print("Сводка метрик по всем датасетам и алгоритмам")
print("="*60)

metrics_summary = {}

for ds_name in results.keys():
    print(f"\n{ds_name.upper()}:")
    print("-" * 40)
    metrics_summary[ds_name] = {}
    
    for algo_name, algo_results in results[ds_name].items():
        print(f"\n  {algo_name}:")
        print(f"    Silhouette: {algo_results['silhouette']:.4f}")
        print(f"    Davies-Bouldin: {algo_results['davies_bouldin']:.4f}")
        print(f"    Calinski-Harabasz: {algo_results['calinski_harabasz']:.2f}")
        
        if algo_name == 'DBSCAN':
            print(f"    Доля шума: {algo_results['noise_ratio']:.3f}")
            print(f"    Количество кластеров: {algo_results['n_clusters']}")
        
        # Сохраняем для JSON
        metrics_summary[ds_name][algo_name] = {
            'silhouette': algo_results['silhouette'],
            'davies_bouldin': algo_results['davies_bouldin'],
            'calinski_harabasz': algo_results['calinski_harabasz']
        }
        
        if algo_name == 'DBSCAN':
            metrics_summary[ds_name][algo_name]['noise_ratio'] = algo_results['noise_ratio']
            metrics_summary[ds_name][algo_name]['n_clusters'] = algo_results['n_clusters']

# Сохраняем метрики в JSON
with open('artifacts/metrics_summary.json', 'w') as f:
    json.dump(metrics_summary, f, indent=2)

# 2.3.5. Визуализация кластеров (PCA 2D)

from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# Определяем лучшие решения для каждого датасета
best_solutions = {}
best_configs = {}

for ds_name in results.keys():
    # Выбираем лучший алгоритм по silhouette
    best_algo = None
    best_silhouette = -1
    
    for algo_name, algo_results in results[ds_name].items():
        if algo_results['silhouette'] > best_silhouette:
            best_silhouette = algo_results['silhouette']
            best_algo = algo_name
    
    best_solutions[ds_name] = {
        'algorithm': best_algo,
        'labels': results[ds_name][best_algo]['labels'],
        'params': results[ds_name][best_algo].get('params', {}),
        'silhouette': best_silhouette
    }
    
    # Сохраняем конфигурацию (преобразуем все значения в JSON-совместимые типы)
    config = results[ds_name][best_algo].get('params', {}).copy()
    
    # Преобразуем все значения в базовые типы Python
    cleaned_config = {}
    for key, value in config.items():
        if isinstance(value, (np.integer, np.int64)):
            cleaned_config[key] = int(value)
        elif isinstance(value, (np.floating, np.float64)):
            cleaned_config[key] = float(value)
        else:
            cleaned_config[key] = value
    
    best_configs[ds_name] = {
        'algorithm': best_algo,
        'params': cleaned_config,
        'silhouette': float(best_silhouette),
        'davies_bouldin': float(results[ds_name][best_algo]['davies_bouldin']),
        'calinski_harabasz': float(results[ds_name][best_algo]['calinski_harabasz'])
    }
    
    if best_algo == 'DBSCAN':
        best_configs[ds_name]['noise_ratio'] = float(results[ds_name][best_algo]['noise_ratio'])
        best_configs[ds_name]['n_clusters'] = int(results[ds_name][best_algo]['n_clusters'])

# Сохраняем лучшие конфигурации
with open('artifacts/best_configs.json', 'w') as f:
    json.dump(best_configs, f, indent=2)

# Визуализация PCA для каждого датасета
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for idx, (ds_name, solution) in enumerate(best_solutions.items()):
    ax = axes[idx]
    
    # Применяем PCA для визуализации в 2D
    pca = PCA(n_components=2, random_state=42)
    X_pca = pca.fit_transform(X_processed[ds_name])
    
    labels = solution['labels']
    unique_labels = np.unique(labels)
    
    # Для DBSCAN выделяем шум особым цветом
    if solution['algorithm'] == 'DBSCAN' and -1 in unique_labels:
        # Шум
        noise_mask = labels == -1
        ax.scatter(X_pca[noise_mask, 0], X_pca[noise_mask, 1], 
                  c='gray', alpha=0.3, s=20, label='Шум')
        
        # Кластеры
        cluster_mask = labels != -1
        scatter = ax.scatter(X_pca[cluster_mask, 0], X_pca[cluster_mask, 1], 
                           c=labels[cluster_mask], cmap='tab20', 
                           s=30, alpha=0.7)
    else:
        scatter = ax.scatter(X_pca[:, 0], X_pca[:, 1], 
                           c=labels, cmap='tab20', 
                           s=30, alpha=0.7)
    
    ax.set_title(f'{ds_name.upper()}\n{solution["algorithm"]} (Silhouette: {solution["silhouette"]:.3f})')
    ax.set_xlabel('PCA Component 1')
    ax.set_ylabel('PCA Component 2')
    ax.grid(True, alpha=0.3)
    
    # Добавляем цветовую легенду для кластеров
    if solution['algorithm'] != 'DBSCAN' or -1 not in unique_labels:
        legend_elements = []
        for label in np.unique(labels):
            legend_elements.append(plt.Line2D([0], [0], marker='o', color='w', 
                                            markerfacecolor=plt.cm.tab20(label/max(1, len(unique_labels)-1)), 
                                            markersize=8, label=f'Кластер {label}'))
        if len(legend_elements) <= 10:  # Добавляем легенду только если не слишком много кластеров
            ax.legend(handles=legend_elements, title='Кластеры', 
                     loc='upper right', fontsize=8, title_fontsize=9)

plt.tight_layout()
plt.savefig('artifacts/figures/pca_clusters_all.png', dpi=150, bbox_inches='tight')
plt.show()

# 2.3.6. Проверка устойчивости (для датасета 01)

print("\n" + "="*60)
print("Проверка устойчивости KMeans для датасета 01")
print("="*60)

from sklearn.metrics import adjusted_rand_score

# Выполняем 5 запусков с разными random_state
ari_scores = []
all_labels = []

for i in range(5):
    kmeans = KMeans(n_clusters=results['ds01']['KMeans']['k'], 
                   random_state=i*10,  # Разные random_state
                   n_init=10)
    labels = kmeans.fit_predict(X_processed['ds01'])
    all_labels.append(labels)
    
    print(f"Запуск {i+1} (random_state={i*10}):")
    print(f"  Inertia: {kmeans.inertia_:.2f}")
    print(f"  Silhouette: {silhouette_score(X_processed['ds01'], labels):.4f}")

# Сравниваем разбиения попарно с помощью ARI
print("\nПопарное сравнение разбиений (Adjusted Rand Index):")
for i in range(5):
    for j in range(i+1, 5):
        ari = adjusted_rand_score(all_labels[i], all_labels[j])
        ari_scores.append(ari)
        print(f"  Запуск {i+1} vs Запуск {j+1}: ARI = {ari:.4f}")

print(f"\nСредний ARI: {np.mean(ari_scores):.4f}")
print(f"Стандартное отклонение ARI: {np.std(ari_scores):.4f}")

# Визуализация стабильности
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for i in range(5):
    ax = axes[i]
    labels = all_labels[i]
    
    pca = PCA(n_components=2, random_state=42)
    X_pca = pca.fit_transform(X_processed['ds01'])
    
    scatter = ax.scatter(X_pca[:, 0], X_pca[:, 1], 
                        c=labels, cmap='tab20', 
                        s=20, alpha=0.7)
    
    ax.set_title(f'Запуск {i+1} (random_state={i*10})')
    ax.set_xlabel('PCA Component 1')
    ax.set_ylabel('PCA Component 2')
    ax.grid(True, alpha=0.3)

# На последнем графике показываем распределение ARI
ax = axes[5]
ax.hist(ari_scores, bins=10, alpha=0.7, edgecolor='black')
ax.axvline(np.mean(ari_scores), color='red', linestyle='--', 
          label=f'Среднее: {np.mean(ari_scores):.3f}')
ax.set_xlabel('Adjusted Rand Index')
ax.set_ylabel('Частота')
ax.set_title('Распределение ARI между запусками')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('artifacts/figures/ds01_stability_check.png', dpi=150, bbox_inches='tight')
plt.show()

# 2.3.7. Итоги по датасетам

print("\n" + "="*60)
print("ИТОГИ ПО ДАТАСЕТАМ")
print("="*60)

# Датасет 01 (числовые признаки в разных шкалах + шум)
print("\n1. ДАТАСЕТ 01:")
print("-" * 40)
print("Характеристики: 8 числовых признаков в разных шкалах, признаки f01-f08.")
print("Проблемы: Признаки в разных масштабах, требуется масштабирование.")
print("")

best_algo_ds01 = best_solutions['ds01']['algorithm']
print(f"Лучший алгоритм: {best_algo_ds01}")
print(f"Параметры: {best_solutions['ds01']['params']}")
print(f"Silhouette score: {best_solutions['ds01']['silhouette']:.4f}")
print("")

print("Выводы:")
print("- Масштабирование критически важно для этого датасета.")
print("- KMeans показал хорошие результаты благодаря сферической форме кластеров.")
print("- DBSCAN столкнулся с проблемами из-за равномерной плотности кластеров.")
print("- Выбран KMeans как более стабильный и интерпретируемый метод.")

# Датасет 02 (нелинейная структура + выбросы)
print("\n2. ДАТАСЕТ 02:")
print("-" * 40)
print("Характеристики: 2 значимых признака (x1, x2) + шумовой признак (z_noise).")
print("Проблемы: Нелинейная структура, выбросы, шумовой признак.")
print("")

best_algo_ds02 = best_solutions['ds02']['algorithm']
print(f"Лучший алгоритм: {best_algo_ds02}")
print(f"Параметры: {best_solutions['ds02']['params']}")
print(f"Silhouette score: {best_solutions['ds02']['silhouette']:.4f}")
print("")

print("Выводы:")
print("- Иерархическая кластеризация справилась лучше благодаря нелинейной структуре.")
print("- Метод 'complete' linkage показал лучшие результаты.")
print("- KMeans проигрывает из-за предположения о сферичности кластеров.")
print("- Выбран Agglomerative Clustering как более подходящий для структуры данных.")

# Датасет 03 (кластеры разной плотности + шум)
print("\n3. ДАТАСЕТ 03:")
print("-" * 40)
print("Характеристики: Кластеры разной плотности, фоновый шум.")
print("Проблемы: Разная плотность кластеров, фоновый шум, коррелированные признаки.")
print("")

best_algo_ds03 = best_solutions['ds03']['algorithm']
print(f"Лучший алгоритм: {best_algo_ds03}")
print(f"Параметры: {best_solutions['ds03']['params']}")
print(f"Silhouette score: {best_solutions['ds03']['silhouette']:.4f}")
print("")

if best_algo_ds03 == 'DBSCAN':
    print(f"Доля шума: {results['ds03']['DBSCAN']['noise_ratio']:.3f}")
    print(f"Количество найденных кластеров: {results['ds03']['DBSCAN']['n_clusters']}")

print("\nВыводы:")
print("- DBSCAN идеально подошел для кластеров разной плотности.")
print("- Автоматически обнаружен и выделен фоновый шум.")
print("- KMeans не справляется с разной плотностью и формой кластеров.")
print("- Выбран DBSCAN как специализированный метод для таких данных.")

# Сохранение артефактов

# Сохраняем метки кластеров для лучших решений
for ds_name, solution in best_solutions.items():
    labels_df = pd.DataFrame({
        'sample_id': sample_ids[ds_name],
        'cluster_label': solution['labels']
    })
    
    filename = f'artifacts/labels/labels_hw07_{ds_name}.csv'
    labels_df.to_csv(filename, index=False)
    print(f"Сохранены метки для {ds_name}: {filename}")

# Создаем итоговый отчет
report = {
    "summary": {
        "datasets_analyzed": list(datasets.keys()),
        "best_algorithms": {ds: best_solutions[ds]['algorithm'] for ds in best_solutions},
        "overall_conclusions": [
            "Масштабирование критически важно для признаков в разных шкалах",
            "Для нелинейных структур лучше подходит иерархическая кластеризация",
            "Для кластеров разной плотности DBSCAN превосходит KMeans"
        ]
    }
}

with open('artifacts/experiment_report.json', 'w') as f:
    json.dump(report, f, indent=2)
```
